## Parte 2: Carregando e Inspecionando (A sua vez!)

Utilize a célula abaixo para usar o Pandas e carregar o `vgsales_sample.csv`. Guarde o resultado numa variável chamada `df` e use as funções `head()` ou `info()` para entender as colunas.

## Parte 4: Agrupamento em Ciência de Dados

Nós queremos saber o TOTAL de vendas (`Global_Sales`) de CADA publicadora (`Publisher`).
No Pandas, há uma função mágica para isso chamada `groupby()` (Agrupar por). Ela funciona exatamente equivalente a uma Tabela Dinâmica do Excel.

```python
# Descomente e rode caso tenha nomeado a tabela de 'df'
vendas_por_empresa = df.groupby('Publisher')['Global_Sales'].sum()

# Isso vai criar uma Série contendo os autores e o total de vendas.
# Para ver as que mais venderam no topo, nós as ordenamos do maior para o menor:
vendas_por_empresa.sort_values(ascending=False)
```
